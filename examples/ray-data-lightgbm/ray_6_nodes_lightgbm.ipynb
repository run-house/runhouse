```python
import runhouse as rh 
num_nodes = 6
cluster_name = f"rh-{num_nodes}-dask-high-mem-gcp"

# The environment for the remote cluster
img = rh.Image("dask-img").pip_install
    [
    #    "dask[distributed,dataframe]",
    #    "dask-ml",
        "ray[data]",
        "gcsfs",
        "pyarrow",
        "scikit-learn",
        "lightgbm",
        "lightgbm_ray",
        "runhouse[server]"
    ],
)

cluster = rh.compute(
    name=cluster_name,
    instance_type="n2-highmem-8",
    num_nodes=num_nodes,
    provider="gcp",
    region="us-east1",
    use_spot=True, 
    autostop_mins=120,
    image=img,
    launcher='den', 
).up_if_not()
cluster.restart_server()
```

```python
cluster.ssh_tunnel(local_port=8265,remote_port=8265)
```

```python
data_path = "gs://rh-demo-external/bitcoin-2024/"
write_data_path = "gs://rh-demo-external/preprocessed_bitcoin-2024/"
date_column = "block_timestamp"
features = ['is_coinbase', 'input_count','output_count','virtual_size','version']
target_var = ['size']
```

```python
preprocess_data = rh.function(preprocess_data).to(cluster, name = 'preprocess').distribute("ray")
preprocessed_columns = preprocess_data(data_path, write_data_path, date_column, columns_to_retain = features+target_var)
```

```python
cluster.restart_server() 
from lightgbm_ray_fns import preprocess_data, train_model
train_model = rh.function(train_model).to(cluster).distribute("ray")    
model = train_model(write_data_path, 'size', test_size = 0.3, n_jobs = num_nodes)
```
